```python
def getreturn(symbol):
    symbol_frame = df[df.symbol == symbol]
    symbol_frame = symbol_frame.sort_values('time')
    symbol_frame = symbol_frame.set_index('time')
    return symbol_frame

# La función getreturn toma un símbolo como argumento y realiza las siguientes operaciones:
# 1. Filtra el DataFrame df para obtener únicamente las filas que corresponden al símbolo especificado.
# 2. Ordena las filas filtradas por la columna 'time' en orden ascendente.
# 3. Establece la columna 'time' como el índice del DataFrame filtrado y ordenado.
# 4. Devuelve el DataFrame resultante.
    
```
